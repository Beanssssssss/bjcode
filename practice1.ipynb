```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Example Plot 1 - PyPlot API style

```python
y = np.array([10, 20, 30])
plt.plot(y)
plt.show()
```

```python
x = np.array([5, 10, 15])
y = np.array([10, 20, 15])
plt.plot(x, y, 'r-o', label = 'Data 1')#r-- 점선
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

Plot Example2 - Object-Oriented Style

```python
hf = plt.figure #새롭게 생성
ha = hf.add_axes([0, 0, 1, 1])
ha.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 초기 데이터
x = np.array([5, 10, 15])
y = np.array([10, 20, 15])

fig, ax = plt.subplots()
line, = ax.plot(x, y, 'r')  # 올바른 방식 (쉼표 추가)
# 새로운 데이터 설정

x_new = np.array([10, 11, 12])
y_new = np.array([10, 10, 10])
line.set_data(x_new, y_new)  # ✅ 정상 작동

# 그래프를 다시 그리기ㄷ
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)  # 2x2 서브플롯 생성

x = np.linspace(0, 10, 100)

axes[0, 0].plot(x, np.sin(x))  # 첫 번째 그래프
axes[0, 1].plot(x, np.cos(x))  # 두 번째 그래프
axes[1, 0].plot(x, np.tan(x))  # 세 번째 그래프


plt.show()
fig
```

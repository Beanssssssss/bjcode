```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Example Plot 1 - PyPlot API style

```python
y = np.array([10, 20, 30])
plt.plot(y)
plt.show()
```

```python
x = np.array([5, 10, 15])
y = np.array([10, 20, 15])
plt.plot(x, y, 'r-o', label = 'Data 1')#r-- 
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

Plot Example2 - Object-Oriented Style

```python
hf = plt.figure # 
ha = hf.add_axes([0, 0, 1, 1])
ha.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  
x = np.array([5, 10, 15])
y = np.array([10, 20, 15])

fig, ax = plt.subplots()
line, = ax.plot(x, y, 'r')  #   ( )
#   

x_new = np.array([10, 11, 12])
y_new = np.array([10, 10, 10])
line.set_data(x_new, y_new)  #   

#   
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)  # 2x2  

x = np.linspace(0, 10, 100)

axes[0, 0].plot(x, np.sin(x))  #   
axes[0, 1].plot(x, np.cos(x))  #   
axes[1, 0].plot(x, np.tan(x))  #   


plt.show()
fig
```

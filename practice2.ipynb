```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Multiple Line Graphs - PyPlot Style

```python
x = np.linspace(0, 2, 100)
y1 = 0.5 * x
y2 = 0.5 * x ** 2
y3 = 0.5 * x ** 3
```

```python
plt.plot(x, y1, label = 'linear')
plt.plot(x, y2, label = 'quadratic')
plt.plot(x, y3, label = 'cubric')
plt.legend()
plt.ylabel('y')
plt.xlabel('x')
plt.title('graphs')
plt.show()
```

Multiple Line Graphs - OOP Style

```python
hf, ha = plt.subplots()
hp1, = ha.plot(x, y1)
hp2, = ha.plot(x, y2)
hp3, = ha.plot(x, y3)
ha.set_ylabel('Y')
ha.set_xlabel('X')
ha.set_title('Graphs')
hp1.set_label('liner')
hp2.set_label('quadratic')
hp3.set_label('cubric')
ha.legend()
plt.show()
```

```python
hp1.set_color('k')
hp2.set_color('b')
hp3.set_color('r')
ha.legend()
ha.grid()
hf
```
